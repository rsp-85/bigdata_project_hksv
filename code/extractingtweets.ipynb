## Extracting tweets

```python
def get_ids(tweet_id_list, start, end):
    tweet_ids = ''
    tweet_id_list1 = tweet_id_list[start:end]
    for r in tweet_id_list1:
        if r != tweet_id_list1[-1]:
            tweet_ids += str(r)+','
        else:
            tweet_ids += str(r)
    return tweet_ids
```

```python
def get_requests(tweet_ids):
    url = 'https://api.twitter.com/2/tweets?ids='+tweet_ids+'&tweet.fields=created_at&expansions=author_id&user.fields=created_at'
    auth = OAuth1(CONSUMER_KEY, CONSUMER_SECRET, OAUTH_TOKEN, OAUTH_TOKEN_SECRET)
    r = requests.get(url, auth=auth)
    response = r.json()
    try:
#     tweets_r = [i['text'] for i in response['data']]
        return response['data']
    except:
        print('no response')
        return 0
```

```python
from tqdm.notebook import tqdm
chunksize = 10000

tweets = []
appended_data = pd.read_csv('out1.csv')
print(appended_data.shape)
for chunk in tqdm(pd.read_csv('../data/full_dataset_clean.tsv', 
                         chunksize=chunksize, 
                         iterator=True)):
    #Preprocessing df
    df1 = chunk[chunk.columns[0]].str.split(expand = True)
    chunk.columns[0]
    df1.rename(columns={0: 'tweet_id', 1: 'tdate', 2: 'ttime', 3: 'tlang', 4: 'tcountry_place'}, inplace = True)
    df2 = df1[df1['tlang'] == 'en']
#     print(df2.shape)
    df2 = df2[df2.tweet_id.isin((set(df2.tweet_id.to_list()) - set(appended_data['id'].to_list())))]
    print(df2.shape)
    
    #Extrating 100 tweets at a time (max limit)
    for i in range(0, df2.shape[0], 100):
        tweet_id_list = list(map(int, df2['tweet_id'].tolist()))
        tweet_ids = get_ids(tweet_id_list, i, i+100)
        try:
            tweets.append(get_requests(tweet_ids))
        except:
            print("NOT APPENDED")
            pass
#     break




```
